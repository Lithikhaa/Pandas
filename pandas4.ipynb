```python
import pandas as pd
```

```python
df = pd.read_csv('/content/LOTR.csv')
df
```

```python
df1 = pd.read_csv('/content/LOTR 2.csv')
df1
```

```python
df.merge(df1,how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df.merge(df1, how = 'inner', on = 'FellowshipID')
```

```python
df.merge(df1, how = 'outer')
```

```python
df.merge(df1, how = 'left')
```

```python
df.merge(df1, how = 'right')
```

```python
df.merge(df1, how = 'cross')
```

***lsuffix and rsuffix*** *are parameters used in methods like merge, join, and combine_first to handle overlapping column names when combining DataFrames*

```python

df.join(df1, on = 'FellowshipID', how = 'outer', lsuffix = '_Left',rsuffix = '_Right')
```

```python
df2 = df.set_index('FellowshipID').join(df1.set_index('FellowshipID'),lsuffix = '_Left',rsuffix = '_Right')
df2
```

```python
df2 = df.set_index('FellowshipID').join(df1.set_index('FellowshipID'),lsuffix = '_Left',rsuffix = '_Right', how = 'outer')
df2
```

```python
pd.concat([df,df1])
```

```python
pd.concat([df,df1],join = 'inner')
```

```python
pd.concat([df,df1],join = 'inner',axis = 1)
```
